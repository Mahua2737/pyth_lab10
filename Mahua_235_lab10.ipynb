```python

import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import re

def validate_form():
    name = name_entry.get()
    email = email_entry.get()
    phone = phone_entry.get()
    gender = gender_var.get()
    day = day_spinbox.get()
    month = month_spinbox.get()
    year = year_spinbox.get()
    condition = condition_text.get("1.0", "end-1c")

    if not name:
        messagebox.showerror("Error", "Please enter your name.")
        return
    if not email:
        messagebox.showerror("Error", "Please enter your email.")
        return
    if not phone:
        messagebox.showerror("Error", "Please enter your phone number.")
        return
    if not gender:
        messagebox.showerror("Error", "Please select your gender.")
        return
   

    name_pattern = re.compile(r'^[A-Za-z\s]*$')
    email_pattern = re.compile(r'^[\w\.-]+@[\w\.-]+\.\w+$')
    phone_pattern = re.compile(r'^\d{10}$')

    if not name_pattern.match(name):
        messagebox.showerror("Error", "Invalid name.")
        return
    if not email_pattern.match(email):
        messagebox.showerror("Error", "Invalid email.")
        return
    if not phone_pattern.match(phone):
        messagebox.showerror("Error", "Invalid phone number.")
        return

    messagebox.showinfo("Success", "Form submitted successfully!")

root = tk.Tk()
root.title("Mental Health Checker")
root.geometry('300x400')
root.config(bg='#FFFF00')

name_label = tk.Label(root, text="Name:")
# name_label.grid(row=0,column=0,sticky=tk.E)
name_label.pack()

name_entry = tk.Entry(root)
# name_entry.grid(row=0,column=1)
name_entry.pack()

email_label = tk.Label(root, text="Email:")
# email_label.grid(row=1,column=0,sticky=tk.E)
email_label.pack()

email_entry = tk.Entry(root)
# email_entry.grid(row=1,column=1,sticky=tk.E)
email_entry.pack()

phone_label = tk.Label(root, text="Phone:")
phone_label.pack()

phone_entry= tk.Entry(root)
phone_entry.pack()

gender_label = tk.Label(root, text="Gender:")
gender_label.pack()

genders = ["Male", "Female", "Other"]

gender_var = ttk.Combobox(root, values=genders)
gender_var.pack()

# gender_var.bind("<<ComboboxSelected>>", handle_selection)

day_label = tk.Label(root, text="Day:")
day_label.pack()

day_spinbox = tk.Spinbox(root, from_=1, to=31)
day_spinbox.pack()

month_label = tk.Label(root, text="Month:")
month_label.pack()

month_spinbox = tk.Spinbox(root, values=("January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"))
month_spinbox.pack()

year_label = tk.Label(root, text="Year:")
year_label.pack()

year_spinbox = tk.Spinbox(root, from_=2000, to=2022)
year_spinbox.pack()

# Create and place radio buttons
condition_label = tk.Label(root, text="Condition:")
condition_label.pack()

condition_var = tk.StringVar()
condition_var.set("Good")

good_radio = tk.Radiobutton(root, text="Good", variable=condition_var, value="Good")
good_radio.pack()
average_radio = tk.Radiobutton(root, text="Average", variable=condition_var, value="Average")
average_radio.pack()
poor_radio = tk.Radiobutton(root, text="Poor", variable=condition_var, value="Poor")
poor_radio.pack()

# Create and place a text box for condition details
details_label = tk.Label(root, text="Condition Details:")
details_label.pack()

condition_text = tk.Text(root, height=5, width=30)
condition_text.pack()
submit_button = tk.Button(root, text="Submit", command=validate_form)
submit_button.pack()
root.mainloop()
```
